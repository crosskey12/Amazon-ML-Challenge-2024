# Questions

```python
#List of questions for each entity type

questions = {
        "depth":  "What is the depth of the item shown in the image? ",
        "height": "What is the height of the item depicted in the image? ",
        "item_volume":  "What is the total volume of the item shown in the image? ",
        "item_weight": "What is the weight of the item shown in the image? ",
        "maximum_weight_recommendation": "What is the maximum weight recommendation for the item in the image? ",
        "voltage": "What is voltage of the item shown in the image? ",
        "wattage": "What is power of the item shown in the image? ",
        "width": "What is the width of the item in the image? "
    }
```

# Initialize Model

```python
# # test.py
import torch
from PIL import Image
from transformers import AutoModel, AutoTokenizer

#Loading the models and tokenizers
model = AutoModel.from_pretrained('openbmb/MiniCPM-Llama3-V-2_5-int4', trust_remote_code=True)
tokenizer = AutoTokenizer.from_pretrained('openbmb/MiniCPM-Llama3-V-2_5-int4', trust_remote_code=True, max_new_tokens=64)
model.eval()
```

# Pipeline

```python
import numpy as np
import pandas as pd
from PIL import Image
from io import BytesIO
import io
import time
import requests
import cv2

#Function downloads image from the url and saves it as PIL.Image file
def download_image(image_url):
    # Send a GET request to the image URL
    response = requests.get(image_url)
    
    if response.status_code == 200:
        # Convert the response content to a numpy array
        image_array = np.frombuffer(response.content, np.uint8)

        # Decode the image array to an image
        image_cv = cv2.imdecode(image_array, cv2.IMREAD_COLOR)

        image_rgb = cv2.cvtColor(image_cv, cv2.COLOR_BGR2RGB)

        # Convert the NumPy array to a PIL image
        image_pil = Image.fromarray(image_rgb)
        
    return response.status_code, image_pil
        
```

```python
#Reading the test.csv file containing image links, processing each image by sending questions to a model, 
# and saves the predictions along with their indices to a new CSV file.


name = 'sample_test.csv' #NAME OF THE FILE SHOULD BE like file_split_100.csv (NOTE: .csv must be present)
input_file_path = f'/kaggle/input/{name}' 
output_file_path = f'output_{name}'

# Load input CSV
df = pd.read_csv(input_file_path)

# Initialize an empty output DataFrame
df_output = []

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    
    start = time.time()

    # Fetch the image
    url = row["image_link"]
    code, image = download_image(url)
    
    if(code == 200):
        # Set the question
        question = "Keep the answer vey short. " + questions[row['entity_name']] + "Give only number and unit as output. Give the answer in format {number unit}. If there are multiple units then provide one of them."
        msgs = [{'role': 'user', 'content': question}]

        res = model.chat(
            image=image,
            msgs=msgs,
            context=None,
            tokenizer=tokenizer,
            temperature=0.7
        )

        # Extract the prediction from the response
        prediction = res # Assuming the result contains a field 'output'
        
    else:
        prediction = ""
    
    # Append the index and prediction to the output DataFrame
    df_output.append({'index': row['index'], 'prediction': prediction})
    
    end = time.time()

    #Logging
    logs = {
        'index': row['index'],
        'prediction': prediction,
        'time': end-start
    }
    
    print(f"Done: {logs}")

# Create DataFrame
df = pd.DataFrame(df_output)

# Save DataFrame to CSV
df.to_csv(output_file_path, index=False)
```
